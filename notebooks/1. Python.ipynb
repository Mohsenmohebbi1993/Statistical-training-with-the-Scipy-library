Fibonacci   
The Fibonacci sequence is a famous numerical sequence in mathematics in which each number is equal to the sum of the two preceding numbers.

%time and time of run

```python
def fib(n):
    f0, f1 = 0, 1
    # one list by n number 1
    f = [1] * n
    for i in range(1, n):
        f[i] = f0 + f1
        f0, f1 = f1, f[i]
        # print(f)
    return f
a = fib(10)
print(f"fibonacci = {a}")


%time
```

random walker max distance

```python
def random_walker_max_distance(M, N):
    traj = [np.random.randn(M).cumsum() for _ in range(N)]
    return np.max(np.abs(traj))

%prun random_walker_max_distance(400, 10000)
```

```python
class QubitHamiltonian():
    def __init__(self, epsilon, delta):
        self.epsilon = epsilon
        self.delta = delta
        
    def _repr_latex_(self):
        return "$ \hat{H} = -%0.2f\hat{\sigma}_z-%0.2f\hat{\sigma}_x $" % \
            (self.epsilon/2, self.delta/2)
```
